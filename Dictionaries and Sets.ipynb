# Dictionaries

```python
#A dictionary consists of keys and values. It is helpful to compare a dictionary to a list. 
#Instead of being indexed numerically like a list, dictionaries have keys. 
#These keys are the keys that are used to access values within a dictionary.
```

```python
Dict={"key1": 1, "key2":[3,3,3], "key3": (4,4), "key4":10, "(0:1)":2}
# A key can be a string, can also be any immutable object such as a tuple 
```

```python
Dict
```

```python
Dict["key1"]
```

```python
release_year_dict = {"Thriller": "1982", "Back in Black": "1980", \
                    "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
```

```python
#like a list, a dictionary holds a sequence of elements. 
#Each element is represented by a key and its corresponding value. 
#Dictionaries are created with two curly braces containing keys and values separated by a colon.
#For every key, there can only be one single value, however, multiple keys can hold the same value. 
#Keys can only be strings, numbers, or tuples, but values can be any data type.
```

```python
#You can retrieve the values based on the names:
release_year_dict['Thriller'] 
```

```python
#we can  retrieve the keys of the dictionary using the method keys():
release_year_dict.keys()
```

```python
#we can retrieve the values using the method values():
release_year_dict.values()
```

```python
#to add (append) an entry:
release_year_dict['Graduation'] = '2007'
```

```python
release_year_dict
```

```python
#delete an entry\element:
del(release_year_dict['Thriller'])
del(release_year_dict['Graduation'])
release_year_dict
```

```python
#I can verify if the key is in the dictionary

'The Bodyguard' in release_year_dict
```

## SETS:

A set is a unique collection of objects in Python. Set is between a pair of curly brackets {}. Python will automatically remove duplicate items.
*A set could not be with duplicated elements*

```python
set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
set1
```

```python
#Convert list to a set:
album_list = [ "Michael Jackson", "Thriller", 1982, "00:42:19", \
              "Pop, Rock, R&B", 46.0, 65, "30-Nov-82", None, 10.0]
album_set = set(album_list)             
album_set
```

```python
music_genres = set(["pop", "pop", "rock", "folk rock", "hard rock", "soul", \
                    "progressive rock", "soft rock", "R&B", "disco"])
music_genres
```

Set Operations (Set can be changed):

```python
A = set ({"dog", "cat", "bird"})
A
```

```python
#to ADD an element:
A.add("elephant")
A
```

```python
#to REMOVE:
A.remove("elephant")
A
```

```python
# Verify if the element is in the set:
"dog" in A
```

### Sets logic operations:

```python
album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
```

```python
album_set1, album_set2
```

```python
#when an element is contained in both sets, the common element will appear in the intersection command and only the common
```

```python
intersection = album_set1 & album_set2
intersection
```

```python
#another way for intersection
album_set1.intersection(album_set2)
```

```python
#You can find all the elements that are only contained in album_set1 using the difference method:
album_set1.difference(album_set2)
```

```python
album_set2.difference(album_set1)
```

```python
# Find the union of two sets

album_set1.union(album_set2)
```

```python
#you can check if a set is a superset or subset of another set,
#respectively, like this:
set(album_set1).issuperset(album_set2)    
```

```python
set(album_set2).issubset(album_set1)     
```

```python
A={1, 2, 333, "rinat"}
B={2, "rinat"}
set(B).issubset(A)
```
